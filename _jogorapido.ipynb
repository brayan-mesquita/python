```python
class Pessoa():
    def __init__(self, nome, idade):
        self.nome = nome
        self.idade = idade
        self._protegido = ["protegido", "protegido_final"] #acesivel por dentro e fora da classe, identificado algo importante que nao deve ser alterado
        self.__privado = ["privado"] #nao acessivel por fora, apenas por funcao ou return, e praticamente imutavel
        self.a = 1
        self.b = 2
    def dizer(self):
        return self.__privado[-1], self._protegido[-1], self.a + self.b

brayan = Pessoa("brayan", 20)
brayan.dizer()
```
