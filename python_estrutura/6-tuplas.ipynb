Você pode usar tuplas para transmitir a qualquer pessoa que esteja lendo seu código que você não pretende que essa sequência de valores seja alterada.

```python
senhas = (123, 456, 789, 'brayan')
print(senhas)
print(senhas[-1])
nome = ('miguel',)
print(type(nome))
nome = ('miguel')
print(type(nome))
```

##conversão para tupla

```python
nome = 'brayan'#convertendo string
nome = tuple(nome)
nome
```

```python
lista = [1, 2, 3, [1, 2, 3], 'brayan']#convertendo lista
print(lista)
lista [-1] = 'miguel'
lista = tuple(lista)
print(lista)
lista = list(lista) + ['patricia']
print(lista)
```
