# Itertools

## combinations

```python
from itertools import combinations
```

```python
lista_nome = ['brayan', 'miguel', 'sophia', 'patricia', 'rosa', 'mesquita']
for combination in combinations(lista_nome[0:4], 2):
    print(combination)
```

```python
lista_numeros = [1, 2, 3, 2]
for combination in combinations(lista_numeros, 3):
    print(combination)
```

Nesse caso nos é retornado as possíveis combinações dos elementos, mas repare que sem haver repetições, independentemente da ordem dos elementos

## permutations

Pagina 532 - 6 em 1 - Coleção Python do ZERO às Redes Neurais Artificiais

Como visto no tópico anterior, realizar combinações pode meio de combinations( ) é eficiente para alguns contextos, porém possui a grande limitação de não retornar as repetições de combinações, o que pode nos ser útil em certos casos.


```python
from itertools import permutations
```

```python
lista_nome = ['brayan', 'miguel', 'sophia', 'patricia', 'rosa', 'mesquita']
for combination in permutations(lista_nome[0:3], 2):
    print(combination)
```

```python
lista_numeros = [1, 2, 3, 2]
for combination in permutations(lista_numeros[0:3], 2):
    print(combination[0] * combination[1],end=' ')
print('\n')
for combination in permutations(lista_numeros, 3):
    print(combination, end=' ')
```

Executando esse bloco, agora temos um retorno mais interessante, mostrando todas as possíveis combinações entre os elementos, inclusive as combinações que se repetem quando apenas invertemos a ordem dos mesmos.

## product
esta funcao permite a combinação de um elemento com ele mesmo, e nesses casos isso é possível por meio da função product, ao contrario das anteriores.

```python
from itertools import product
```

```python
lista_nome = ['brayan', 'miguel', 'sophia', 'patricia', 'rosa', 'mesquita']
for combination in product(lista_nome,repeat=2):
    print(combination, end=' ')
```
