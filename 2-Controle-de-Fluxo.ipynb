## Condição if e elif

```python
senha, usuario = 123, "Maria"
login = input("Digite seu nome de usuario: ")
if login == usuario:
  print(f"Ola {usuario}")
  password = input("Digite sua senha: ")
  if int(password) == senha:
    print("Logado com sucesso!")
  else:
    print("Senha errada")
else:
  print("Voce não esta autorizado")   
```

```python
nome, idade = "Brayan", 30
if nome == "Brayan" and idade == 30:
  print(f"Ola {nome}")
elif idade != 30:
  print("Você nao tem 30")
else:
  print("Voce nao e Brayan")
```

##break

As instruções break e as instruções continue são usadas dentro de loops



```python
nome = "Brayan"
alguem = ""
while nome != alguem:
  alguem = input("Seu nome: ")
  if nome != alguem:
    print("Tente novamente!")
  if alguem == "naosei":
    break
print(f"Ola, {alguem}")
```

```python
while True:
  nome = input('Seu nome! ')
  if nome ==  'miguel':
    print('ola, miguel!')
    break
```

##continue

```python
while True:
  nome = input("Seu nome? ")
  if nome != "Brayan":
    continue
  print(f"Olá {nome}")
  senha = input("Qual a senha?")
  if senha != "123":
    continue
  break
```

```python
n = 0
while True:
  if (n % 4) == 0 and n != 0:
    print(f"O numero {n} é divisivel por 4, sendo o resultado: {int(n/4)}.")
  n += 1
  if n == 50:
    break
```

#Encenrrando execução

##sys.exit()
terminando um programa antecipadamente com sys.exit() O último conceito de controle de fluxo a ser abordado é como encerrar o programa. Isso sempre acontece se a execução do programa atingir o final das instruções

```python
import sys
while True:
  print('Type exit to exit.')
  response = input()
  if response == 'exit':
    sys.exit()
  print('You typed ' + response + '.')
```

#Operadores Booleanos

O computador avaliará primeiro a expressão da esquerda e, em seguida, avaliará a expressão da direita. Quando souber o valor booleano de cada um, ele avaliará toda a expressão até um valor booleano

```python
print((5>3) and (6>8) or (5 == 5)) #verdadeiro
print((5>3) and ((6>8) or (5 != 5))) #falso
print(False and True)#falso
print(True or not(False and True))#verdadeiro
print(not True)#falso
```
