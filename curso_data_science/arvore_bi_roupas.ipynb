```python
class Filhos:
    def __init__(self):
        self.vertices = list()
    def inserir_dado(self, dado):
        vertice_novo = Vertice(dado)
        self.vertices.append(vertice_novo)
        return vertice_novo
    def imprimir_vertice_novo(self):
        print(self.vertice_novo)
    def imprimir_pre_ordem(self):
        """
        Percorre a árvore em pré ordem (esquerda, direita, vértice)
        e imprime o dado do vértice
        :return: None
        """
        for vertice in self.vertices:
            vertice.imprimir_pre_ordem()
```

```python
class Vertice:
    def __init__(self, dado):
        self.dado = dado
        self.filhos = None
    def __str__(self):
        return str(self.dado)
    def repre_parenteses(self):
        pass
    def repre_recuo(self):
        pass
    def inserir_filho(self, dado):
        if self.filhos == None:
            self.filhos = Filhos()
        return self.filhos.inserir_dado(dado)
    def imprimir_pre_ordem(self):
        print(self)
        if self.filhos:
            self.filhos.imprimir_pre_ordem()
    def imprimir_dado(self):
        print(self.dado)
    def imprimir_filhos(self):
        print(self.filhos.imprimir_vertice_novo)
```

```python
raiz = Vertice("raiz")
masculino = raiz.inserir_filho("masculino")
feminino = raiz.inserir_filho("feminino")
calcas = masculino.inserir_filho("calças")
saias = feminino.inserir_filho("saias")
```

```python
masculino.imprimir_pre_ordem()
```
