```python
class Dados():
    def __init__(self):
        self.itens = []

    def __repr__(self):#método de retorno de informações da classe
        return str(self.itens)

    def insere(self, valor):
        self.itens.append(f"{valor},  ítem")

    def remover(self, valor=''):
        self.itens.pop()

def main():
    #instanciando classe para dados
    dados = Dados()

    #inserindo valores
    dados.insere('element')
    dados.insere('brayan')
    dados.insere('miguel')

    print(dados)
    #remover valores
    dados.remover()
    dados.remover()

    print(dados)



if __name__ == "__main__":
    main()
```
