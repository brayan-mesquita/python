##pequeno programa utilizndo listas

```python
aniversarios = {'Brayan': 16, 'patricia': 2}
while True:
  nome = input()
  if nome == '':
    break
  elif nome in aniversarios:
    print(f'{nome.capitalize()} seu aniversario e {aniversarios[nome]}')
  else:
    print('Voce nao esta cadastrado, diga seu nome depois dia aniverario')
    nome= input()
    idade= input()
    aniversarios[nome] = idade
    print(f'{nome} Adicionado!')#pequeno programa utilizndo listas
```

##criando dicionários com função dict

```python
idades = dict(brayan=30, miguel=5)
print(idades)
print(idades['brayan'])
print(idades.get('miguel'))
```

##loop em dicionário

```python
telefone = dict(joao=2130, maria=239423, patricia=20390, brayan=209234)
#percorrendo chaves
for i in telefone.keys():
    print(i)
for i in telefone.values():
    print(i)
for i in telefone.items():
    print(i)
```

##removendo itens e adicionando

```python
telefone = dict(joao=2130, maria=239423, patricia=20390, brayan=209234)
print(telefone)
del telefone['joao']
print(telefone)
telefone['miguel'] = 323
print(telefone)
```

##mudar dados dos dicionários e apgar

```python
telefone = dict(joao=2130, maria=239423, patricia=20390, brayan=209234)
print(telefone)
telefone['brayan'] = 999909596
print(telefone)
del telefone['brayan'], telefone['patricia']
print(telefone)
telefone['miguel'], telefone['sophia'], telefone['brayan'] = 111111, 222222, telefone['joao'] *1000
print(telefone)
```

```python
pessoa = {'nome': 'brayan',
            'idade': 30, 
            'formação': ['contador', 'ciência de dados']}
print(pessoa)
pessoa['formação'].append('administração')#adicionando em lista dentro de dicionário
print(pessoa)
pessoa['nome'] = 'miguel'
print(pessoa)
```
